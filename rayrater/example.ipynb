```python
from cf_compact.utils.slam_classes import MapObjectList


def load_map_objects(map_objects_file):
    map_objects = MapObjectList()
    import pickle
    import gzip
    # open pkl.gz file
    def load_pkl_gz(file):
        try:
            # Open the gzip-compressed pickle file in binary read mode
            with gzip.open(file, 'rb') as file:
                data = pickle.load(file)
            return data
        except FileNotFoundError:
            print(f"Error: The file '{pickle_file}' was not found.")
        except pickle.UnpicklingError:
            print(f"Error: The file '{pickle_file}' could not be unpickled.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")
    
    map_objects_dict = load_pkl_gz(map_objects_file)
    print(map_objects_dict)
    print(list(map_objects_dict.keys()))
    map_objects.load_serializable(map_objects_dict['objects'])
    return map_objects
```
